```python
logs_df.loc[logs_df.traffic_source.str.contains('yandex|google'),'source_type'] = 'organic'
```

```python
logs_df.loc[(logs_df.traffic_source.str.contains('paid|email')) &
            (logs_df.region.str.contains('Russia')),'source_type'] = 'ad'
```

```python
logs_df.loc[(logs_df.traffic_source.str.contains('paid|email')) &
            (~logs_df.region.str.contains('Russia')),'source_type'] = 'other'
logs_df.head()
```

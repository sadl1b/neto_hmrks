### Задание 1

В магазин привезли устройства с 3-х разных предприятий.

Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта.

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.

```python
b1 = 0.2
b2 = 0.3
b3 = 0.5
pa_b1 = 0.1
pa_b2 = 0.05
pa_b3 = 0.2
```

Имеем событие A - извлечена продукция высшего сорта

Имеем три гипотезы: 
$B_1$ - выбрали первое предприятие
$B_2$ - выбрали второе предприятие
$B_3$ - выбрали третье предприятие

Гпотезы независимые, в конечном итоге по формуле полной вероятности для несовместных событий получим:
$P(A) = P_{B_1}(A)\cdot P(A)+P_{B_2}(A)\cdot P(A)+P_{B_3}(A)\cdot P(A)$

```python
answer = pa_b1 * b1 + pa_b2 * b2 + pa_b3 * b3
answer
```

## Матрица исходов

Докажу попарную независимость на примере А1 и А2:
Несколько события назввают независимыми, если каждые два из них попарно независимы

Событие А2 будет независимо от события А1, если его условная вероятность от происхождения события А1 равна безусловной вероятности

$P_{A1}(A_2) = P(A_2)$

опишу все возможные элементарные исходы

о - орел

Без доказательства скажу, что во всех трех событиях такая вероятность (считал на бумаге)

Если условная веростность $P_{A1}(A_2)$ будет равна $\frac{1}{2}$ (будем считать, что и со всеми остальными парамитакже), то события являются попарно независимыми

Итак для A1 нам подходят следущие исходы - *ооо, рро, оор, ррр*

Расчитывая А2 *при условии*, что А1 наступило, общее количество исходов будет равно 4 (как все возможные для А1, ведь считаем, что оно может произойти только в таком случае),а вместо благоприятных исходов, которые были выше - *ооо, орр, роо, ооо* нам подойдут только *ооо и ррр* и получаем:

$P_{A_1}(A_2) = \frac{2}{4} = \frac{1}{2}$

Теперь докажем, что они не являются независимыми в совокупности:

События называют независимыми в совокупности, если независсимы каждые два из них и независимы каждое событие и все возможные произведения, поэтому нам нужно найти хотя бы одну комбинацию, которая не выполняет данное условие, например:

$P_{A_1 A_2}(A_3) != P(A_3)$

мы получили 4 возможных исхода - *ооо, рро, оор, ррр*

### Задание 3


Изучите 3 распределения из библиотеки scipy: [Нормальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), [Экспоненциальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html), [Стьюдента](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html). Используя функцию плотности распредления (`pdf`), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

Дополнительная информация [тут](https://pythonguides.com/scipy-stats/).

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import seaborn as sns
```

```python
location = np.linspace(-10,10,5)
scale = np.linspace(-1,5,5)
data = np.linspace(-10,10,100)
```

```python
fig, axs = plt.subplots(5, 1, figsize=(10,15))

for i,value in enumerate(location):
    ax = axs[i]
    ax.plot(data,stats.norm.pdf(data, loc=value, scale=scale[i]))
    ax.set_title(f'Location is {value} and scale is {scale[i]}')

        
for ax in axs.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=25, fontsize=10)
    ax.set_ylabel('Stat value')
    
fig.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(5, 1, figsize=(10,15))
freedom_deg= np.arange(1,10,2)
for i,value in enumerate(location):
    ax = axs[i]
    ax.plot(data,stats.t.pdf(data,freedom_deg[i], loc=value, scale=scale[i]))
    ax.set_title(f'Location is {value} and scale is {scale[i]}')

        
for ax in axs.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=25, fontsize=10)
    ax.set_ylabel('Stat value')
fig.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(5, 1, figsize=(10,15))

for i,value in enumerate(location):
    ax = axs[i]
    ax.plot(data,stats.expon.pdf(data, loc=value, scale=scale[i]))
    ax.set_title(f'Location is {value} and scale is {scale[i]}')

        
for ax in axs.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=25, fontsize=10)
    ax.set_xlim(-5,20)
    ax.set_ylabel('Stat value')
fig.tight_layout()
plt.show()
```

По результатам можно отметить, что location приближен к мат ожиданию, тоесть показывает наиболе частотную часть распределения данных, scale отвечает за растяжение графика и не может быть отрицательным - близко к дисперсии и ско, т-распределение сильно похоже на нормальное, влияние степеней свободы не удалось отразить, экспоненциальное - экспоненциальное распределение что тут сказать, по идее не должно быть отрицательным, но если двигать loc видимо можно

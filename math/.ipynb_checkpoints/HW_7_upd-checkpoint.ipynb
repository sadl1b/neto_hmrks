### Задание

При фиксированном `seed=42` поэкспериментируйте с параметрами алгоритма [differential_evolution](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.differential_evolution.html): strategy, popsize, mutation.

Постройте графики количества итераций (`nit`) оптимизации [функции ackley](https://en.wikipedia.org/wiki/Ackley_function) от значения параметра.

Подробнее о результате выдачи [тут](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.OptimizeResult.html).

```python
import numpy as np
from scipy.optimize import differential_evolution
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-10, 10), (-10, 10)]

result = differential_evolution(ackley, bounds, seed=42)
result
```

```python
mutation_range = np.arange(0.3,1.5,0.2)
popsize_range = np.arange(10,30,5)
strategies = ['best1bin','best1exp','rand1bin','rand1exp','rand2bin','rand2exp','randtobest1bin','randtobest1exp',
            'currenttobest1bin','currenttobest1exp','best2exp','best2bin']
```

```python
df = pd.DataFrame(columns=('strategy', 'mutation','popsize','nit'))
```

```python

for strtg in strategies:
    for p_multiplier in popsize_range:
        for m_multiplier in mutation_range:
            df.loc[len(df)] =[strtg,m_multiplier, p_multiplier ,differential_evolution(ackley, bounds, seed=42, 
                                                                        strategy = strtg,
                                                                        popsize = p_multiplier,
                                                                        mutation = m_multiplier).nit]
```

```python
df
```

```python
df.drop(['strategy','popsize'],axis=1).plot(kind='scatter',x='mutation', y = 'nit')
plt.title('nit dependance over mutation multiplier')
```

```python
df.drop(['strategy','mutation'],axis=1).plot(kind='scatter',x='popsize', y = 'nit')
plt.title('nit dependance over popsize multiplier')
```

```python
df.drop(['mutation','popsize'],axis=1).plot(kind='scatter',x='strategy', y = 'nit')
plt.xticks(rotation=45)
plt.title('nit dependance over method')
```

Не рекомендую всерьез воспринимать все что ниже,я все же не могу вьехать как же мне такую бешеную производную из аналититической превратить в программную, в формуле еще x и y, которые я так понимаю достаются индексирования x[0] и x[1]. В общем оставлю толкьо BFGS, матрицу Гессе не буду даже пробовать считать, надеюсь после замечаний разберусь.

```python
def ack_der(x):
    x0m = x[1: -1]
    x0m_m1 = x[:-2]
    x1m = x[1: -1]
    x1m_m1 = x[:-2]
    der = np.zeros_like(x)
    der[1: -1] = -((0.4*(x0m_m1*x1m_m1*np.exp(-0.14*np.sqrt(x0m**2+x1m**2))))/(x0m**2+x1m**2))+\
                    (2.8*x0m*x1m*np.exp(-0.14*np.sqrt(x0m**2+x1m**2)))/(x0m_m1**2+x1m_m1**2)**1.5+\
                    np.pi**2*np.exp(0.5*np.cos(2*np.pi*x0m) + 0.5*np.cos(2*np.pi*x1m)) +\
                    np.sin(2*np.pi*x0m_m1)*np.sin(2*np.pi*x1m_m1)
    return der
```

```python
from scipy.optimize import minimize

x0 = [0,0]

result_m = minimize(ackley, x0,  method='BFGS', jac=ack_der)
result_m
```
